```python
import numpy as np
import xarray as xr
import pandas as pd
import netCDF4
import glob
import os
```

```python
df=pd.read_csv('claire.csv')
df
```

```python
pdir='/glade/u/home/djk2120/mapp_droughts/cce/params/paramfiles/'
basefile=pdir+'base.nc'
```

```python
v='frac_sat_soil_dsl_init'
ix=df.param==v
ps=df.key_CLM5PPE[ix].values
for p in ps:
    nextnum = 1+int(sorted(glob.glob(pdir+'p*'))[-1].split('/p')[-1].split('.')[0])
    pfile = pdir+'p'+str(nextnum).zfill(2)+'.nc'
    
    pref = '/glade/campaign/asp/djk2120/PPEn11/paramfiles/'+p+'.nc'
    ## CREATE and EDIT the new paramfile
    cmd = 'cp '+basefile+' '+pfile
    os.system(cmd)
    cmd = 'chmod 644 '+pfile
    os.system(cmd)

    dset = netCDF4.Dataset(pfile,'r+')
    x = xr.open_dataset(pref)[v]
    dset[v][:] = x
    dset.close()
```
